```python
# Coefficient matrix A
A_eq = np.array([[2, -3, 1],
                 [1, -1, 2],
                 [3, 1, -1]])

# Constant matrix B
B_eq = np.array([-1, -3, 9])
```

```python
X = np.dot(np.linalg.inv(A_eq), B_eq)
print("Solution [x, y, z] =", X)
```

4.2 Experiment: How Fast is Numpy?
In this exercise, you will compare the performance and implementation of operations using plain Python lists (arrays) and NumPy arrays. Follow the instructions:

1.Element-wise Addition:

‚Ä¢ Using Python Lists, perform element-wise addition of two lists of size 1, 000, 000. Measure and Print the time taken for this operation.

‚Ä¢ Using Numpy Arrays, Repeat the calculation and measure and print the time taken for this operation.

```python
import numpy as np
import time
#here we are declaring the size of the arrays
N = 1_000_000

# Python lists
list1 = list(range(N))
list2 = list(range(N))

# NumPy arrays
arr1 = np.array(list1)
arr2 = np.array(list2)
start_time = time.time()

# Using list comprehension for element-wise addition
list_sum = [list1[i] + list2[i] for i in range(N)]

end_time = time.time()
print("Time taken for Python lists addition: {:.5f} seconds".format(end_time - start_time))
start_time = time.time()

# Using NumPy's vectorized addition
array_sum = arr1 + arr2

#displaying the result of the difference of the time between element wise and numpy additions
end_time = time.time()
print("Time taken for NumPy arrays addition: {:.5f} seconds".format(end_time - start_time))
```

4.Matrix Multiplication
‚Ä¢ Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Measure and print the time taken for this operation.

‚Ä¢ Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000. Measure and print the time taken for this operation.

```python
import numpy as np
import time
import random
N = 1000

# Generate 1000x1000 matrices with random integers 0-10
matrix1 = [[random.randint(0, 10) for _ in range(N)] for _ in range(N)]
matrix2 = [[random.randint(0, 10) for _ in range(N)] for _ in range(N)]
np_matrix1 = np.array(matrix1)
np_matrix2 = np.array(matrix2)
start_time = time.time()

# Initialize result matrix with zeros
result_list = [[0 for _ in range(N)] for _ in range(N)]

# Perform multiplication
for i in range(N):
    for j in range(N):
        for k in range(N):
            result_list[i][j] += matrix1[i][k] * matrix2[k][j]

end_time = time.time()
print("Time taken for Python lists matrix multiplication: {:.5f} seconds".format(end_time - start_time))
start_time = time.time()

result_np = np.dot(np_matrix1, np_matrix2)  # or np_matrix1 @ np_matrix2

end_time = time.time()
print("Time taken for NumPy arrays matrix multiplication: {:.5f} seconds".format(end_time - start_time))
```
